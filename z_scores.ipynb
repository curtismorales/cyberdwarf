```python
import yaml

import pandas as pd
from unidecode import unidecode_expect_ascii

from stats import score_cohort
```

```python
# Import all box scores
main_box_scores = pd.read_csv("all_box_scores.csv")
other_box_scores = pd.read_csv("other_box_scores.csv")
box_scores = pd.concat([main_box_scores, other_box_scores])
box_scores["player"] = box_scores["player"].apply(unidecode_expect_ascii)
box_scores
```

```python
box_scores.columns
```

```python
z_scores = score_cohort(box_scores)
z_scores
```

```python
with open("bubble_rosters.yml") as file:
    bubble_rosters = yaml.safe_load(file)
    
all_players = []
for team, players in bubble_rosters.items():
    all_players += [[team, p] for p in players]

bubble_df = pd.DataFrame(all_players)
bubble_df.columns = ["team", "player"]
```

```python
bubble_box_scores = pd.merge(box_scores, bubble_df, on="player", how="left")
bubble_box_scores["team"] = bubble_box_scores["team_y"]
bubble_box_scores["team"].fillna("out", inplace=True)

bubble_box_scores
```

```python
bubble_zs = score_cohort(bubble_box_scores)
bubble_zs
```

```python
bubble_zs.to_csv("bubble_zs.csv")
```
